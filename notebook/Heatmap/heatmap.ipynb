### Notebook para geração do dataset de heatmaps

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import cv2
```

```python

imagem = plt.imread("../../data/base_recortada/ASC-H/10.png")

intensidade = np.mean(imagem, axis=2)  


cores = ['purple', 'yellow']
limites = [intensidade.min(), intensidade.max()]


heatmap = plt.imshow(intensidade, cmap=plt.cm.colors.ListedColormap(cores), vmin=limites[0], vmax=limites[1])


plt.colorbar(heatmap)


plt.show()
```
